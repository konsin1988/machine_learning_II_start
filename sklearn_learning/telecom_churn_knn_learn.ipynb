```python
import pandas as pd
import numpy as np
import sklearn as sk
import sqlite3

connection = sqlite3.connect('telecom_churn.db')
cursor = connection.cursor()

def select(query):
    try:
        return pd.read_sql_query(query, connection)
    except:
        cursor.executescript(query)
        connection.commit()

```

```python
telecom = pd.read_csv('telecom_churn.csv')
telecom.head()
```

```python
telecom.info()
```

### Divide columns to signs and marks (признаки и метки)

```python
X = telecom.drop(columns='Churn')
y = telecom['Churn']
```

### Divide dataset to train and validation (to test group we don't have anough data)

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25)
```

### Create object that classifies data via knn (k-nearest neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn
```

### Teaching model:

```python
knn.fit(X_train, y_train)
```

### Get the predict value:

```python
preds_valid = knn.predict(X_valid)
preds_valid
```

>After receiving the predictions, we need to check how good they were. Let's use the **accuracy** metric

```python
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
print(f'Accuracy = {accuracy_score(y_valid, preds_valid)}')
print(f'Precision = {precision_score(y_valid, preds_valid)}')
print(f'Recall = {recall_score(y_valid, preds_valid)}')
```

## Выводы
- Серди тех, для кого мы предсказали 1 (т.е. решили, что они уйдут), всего лишь половина (precision) действительно собиралась уйти
- Среди тех, кто действительно ушел, мы нашли только 33% (recall). Т.е. если бы мы даже предложили максимально щедрое предложение, призванное удержать клиента, мы бы смогли удержать только 33%.

> The biggest problem with knn is the curse of dimensionality. When we have <br> a lot of features, the density of training examples in n-dimensional space decreases dramatically.

## сильные и слабые стороны KNN

#### Плюсы:

    Хороший бейзлайн.
    Хорошая интерпретируемость при небольом числе признаков.
    Быстрый на небольших и средних данных.
    Теоретически идеален и не делает никаких предположений о зависимости

#### Минусы:

    Нужно хорошее пространство и метрика, чтобы он работал. Т.е. нужно отбирать признаки, уменьшать размерность данных, стандартизировать признаки итд.
    Если можно сделать какие-то предположения о природе данных, их нужно встраивать в модель через создание новых признаков или измение метрики. Очень сложно и требует опыта.


У нас нет пропусков, что значительно облегчает задачу.

    Стандартизируем признаки
    Переберем метрики и k neighbors.


```python
import pandas as pd
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X = telecom.drop(columns = ['Churn'])
y = telecom['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.25)
```

### Let's standardize the signs

### Нормировать признаки стоит почти всегда. Это очень важно для

-    Метрических алогритмов
-    Линейных алогритмов (мы еще узнаем, что признаки хорошо не только отнормировать, но и как можно сильнее приблизить их распределение к нормальному). Сюда относятся как всевозможные регрессии, так и PCA, который мы используем ниже.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train, y_train)
X_train_norm = scaler.transform(X_train)
X_valid_norm = scaler.transform(X_valid)
```

### Train the model

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_norm, y_train)
preds_valid = knn.predict(X_valid_norm)
```

### Testing the model

```python
print(f'Accuracy = {accuracy_score(y_valid, preds_valid)}')
```
