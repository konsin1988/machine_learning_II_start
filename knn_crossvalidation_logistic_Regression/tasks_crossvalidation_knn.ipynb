```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('forest_dataset.csv')
df.head()
```

```python
df.info()
```

```python
labels = df.iloc[:,-1].values
```

```python
feature_matrix = df.iloc[:,:-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_matrix, labels, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(C=1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
model = LogisticRegression(solver='saga')
# params to search in
param_grid = {
    'C': np.arange(1, 5),
    'penalty': ['l1', 'l2']
}

search = GridSearchCV(model, param_grid, n_jobs=-1, cv=5, refit = True, scoring='accuracy')
search.fit(X_train, y_train)
print(search.best_params_)
```

```python
y_pred = search.best_estimator_.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
params = {
    'n_neighbors': np.arange(1, 10),
    'metric': ['manhattan', 'euclidean'],
    'weights': ['uniform', 'distance']
}
search = GridSearchCV(model, param_grid=params, n_jobs=-1, cv=5, refit = True, scoring='accuracy')

search.fit(X_train, y_train)
print(search.best_params_)
```

```python
y_pred = search.best_estimator_.predict(X_test)
accuracy_score(y_test, y_pred)
```
